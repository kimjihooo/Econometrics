## Basics of Data Visualization Using Matplotlib and Pandas

The <b>pandas.Series.plot</b> and <b>pandas.DataFrame.plot</b> methods provide basic but convenient visualization functionality. They are, in fact, a simple wrapper around <b>matplotlib.pyplot.plot</b> in the matplotlib package. 

```python
import numpy as np
import pandas as pd 

np.random.seed(0)
series = pd.Series(np.random.randint(1, 101, 10))   # from an array with 10 random integers between 1 and 100 
series
```

```python
series.plot(kind="line", grid=True)
```

pandas.Series.plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.html

When drawing a plot on a Pandas series using the <b>plot</b> method, the x axis is the index of the series, while the y axis is its values.

`kind`: str 

The kind of plot to produce:
- line: line plot (default)
- bar: vertical bar plot
- barh: horizontal bar plot
- hist: histogram
- box: boxplot
- kde: Kernel Density Estimation plot
- density: same as ‘kde’
- area: area plot
- pie: pie plot

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randint(1, 101, (10,3)),  # from a 10 x 3 array with random integers between 1 and 100
                  columns=["a", "b", "c"])
df
```

```python
df.plot(kind="line")
```

pandas.DataFrame.plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

When drawing a plot on a Pandas dataframe using the <b>plot</b> method, the x axis is the index of the dataframe, while the y axis is the values of the columns in the dataframe. In this example, there are 3 lines that correspond to the 3 columns.

```python
df[["b", "c"]].plot(kind="line")
```

You can select some of the columns you are interested in. 

```python
import matplotlib.pyplot as plt

df[["b", "c"]].plot(kind="line")
plt.show()                                        # Remove the plotting objects information above the plot.
```

A box plot is a method for graphically depicting groups of numerical data through their quartiles (Q1, Q2, and Q3). The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). The whiskers extend from the edges of box to show the range of the data. The position of the whiskers is set by default to 1.5 * IQR (IQR = Q3 - Q1) from the edges of the box. Outlier points are those past the end of the whiskers.

```python
dfb19[["AB", "H", "2B", "3B", "HR", "BB", "SO"]].plot(kind="box", grid=True, figsize=(15,7))
```

A scatter plot is a two-dimensional data visualization that uses dots to represent the values obtained for two different variables - one plotted along the x axis and the other plotted along the y axis. This kind of plot is useful for visualizing complex correlations between two variables. 

- Positive correlation: as one variable increases so does the other
- Negative Correlation: as one variable increases, the other decreases
- No correlation: there is no apparent relationship between the two variables

- spearman : Spearman rank correlation

## Visualization of Time Series Data 

When visualizing data with timestamps, it is important to convert the data to timeseries data, such that you calculate the row counts by a certain time unit such as day, week, month, or year. 

```python
dft = pd.read_csv("classdata/timeline_cnnbrk.csv", sep="\t")
dft
```

```python
dft.info()
```

```python
dft.created_at = dft.created_at.astype("datetime64")
```

If you set the data types of columns in a dataframe to datetime64, you can do some useful date/time operations on those columns. 

```python
dft
```

```python
dft.info()
```

```python
pd.Series(1, index=dft.created_at)
```

```python
count = pd.Series(1, index=dft.created_at).resample(rule="d").count()     # Count the number of rows per day. 
count.head(10)
```

pandas.Series.resample: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.resample.html

The `rule` argument value takes
- "h": hourly
- "d": daily
- "w": weekly
- "m": monthly
- "y": yearly

```python
count.plot(kind="line", title="Tweet Count Change over Time", figsize=(15,7))
```

```python
count.max()
```

```python
count.idxmax()
```

pandas.Series.idxmax: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.idxmax.html

The <b>idxmax</b> method returns the row index label of the maximum value, whereas the <b>argmax</b> method returns the row index position of the maximum value.

## Interactive Visualization

Interactive visualization allows users to interactively update the current plot by changing its parameter values. Ipywidgets are interactive HTML widgets for Jupyter notebooks and the IPython kernel.

```python
from ipywidgets import widgets, interactive, Layout
```

```python
w = widgets.ToggleButtons(
    description = 'Speed:',
    options = ['Slow', 'Regular', 'Fast'],
    value = 'Slow',
    style = {"description_width": '50px'},
    layout = Layout(width="70%")
)

display(w)
```

ToggleButtons Widget: https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#ToggleButtons

Widgets are eventful Python objects that have a representation in the browser, often as a control like a slider, textbox, etc.

```python
w.value
```

The <b>value</b> attribute of the widget holds the chosen value. 

```python
w_league = widgets.ToggleButtons(
    description='League:',
    options=['AL', 'NL', 'Both'],
    value='Both',
    style = {"description_width": '50px'},
    layout = Layout(width="70%")
)

display(w_league)
```

```python
w_league.value
```

```python
if w_league.value == "Both":                         # If the user chose 'Both',
    df_tmp = dfb19                                   # select the entire dataframe
else:                                                # If the user chose either 'AL' or 'NL',
    df_tmp = dfb19[dfb19.lgID == w_league.value]     # select the rows with the chosen league 

title = "Batting Stats for {}".format(w_league.value)
df_tmp[["AB", "H", "2B", "3B", "HR", "BB", "SO"]].plot(kind="box", title=title, grid=True, figsize=(15,7))
```

### Make it interactive!

```python
w_league = widgets.ToggleButtons(
    description='League:',
    options=['AL', 'NL', 'Both'],
    value='Both',
    style = {"description_width": '50px'},
    layout = Layout(width="70%")
)

def view(league):
    if league == "Both":
        df_tmp = dfb19
    else:
        df_tmp = dfb19[dfb19.lgID == league]
    
    title = "Batting Stats of {}".format(league)
    df_tmp[["AB", "H", "2B", "3B", "HR", "BB", "SO"]].plot(kind="box", title=title, grid=True, figsize=(15,7))

i = interactive(view, league=w_league)  # The value of the widget is passed as a parameter value.
display(i)
```

```python
w = widgets.Dropdown(
    description='Speed:',
    options=['Slow', 'Regular', 'Fast'],
    value='Slow',
    style = {"description_width": '50px'},
    layout = Layout(width="20%")
)

display(w)
```

Dropdown Widget: https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#Dropdown

```python
w.value
```

```python
w_team = widgets.Dropdown(
    description='Team:',
    options = ["All"] + sorted(list(set(dfb19.teamID))),  # a sorted list of unique teamIDs plus All
    value = "All",
    style = {"description_width": '50px'},
    layout = Layout(width="15%")
)

def view(team):
    if team == "All":
        df_tmp = dfb19
    else:
        df_tmp = dfb19[dfb19.teamID == team]

    title = "Batting Stats of {}".format(team)
    df_tmp[["AB", "H", "2B", "3B", "HR", "BB", "SO"]].plot(kind="box", title=title, grid=True, figsize=(15,7))

i = interactive(view, team=w_team)
display(i)
```

```python
w_string = widgets.Text(
    description='String:',
    style = {"description_width": '50px'},
    layout = Layout(width="90%")
)

def view(string):
    mask = dft.text.str.contains(string, case=False, regex=False)
    display(dft[mask])

i = interactive(view, string=w_string)
display(i)
```

Note that you can use interactive visualization using Ipywidgets only in a notebook environment. In other words, you lose the interactive functionality once you convert it to HTML or other formats.
