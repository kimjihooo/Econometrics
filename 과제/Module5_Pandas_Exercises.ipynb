## Exercises for Handling Null Values

Suppose you have `df1`, `df2`, `df3`, `df4`, and `df5`, each of which is a copy of `df`.

```python
df = load_dataset("titanic")
df = df[["survived", "pclass", "sex", "age", "fare"]].sample(n=10, random_state=4)
df.iloc[1, 0] = None
df.iloc[7, 1] = None
df.iloc[3, 4] = None
df.iloc[[5, 9], :] = None

df
```

```python
df1, df2, df3, df4, df5 = df.copy(), df.copy(), df.copy(), df.copy(), df.copy()
```

1\. Select all rows in `df` with any null values.

```python
# Your answer here
mask = df.isnull().any(axis=1)
df[mask]
```

2\. Drop the rows in `df1` that have any null values. Make sure to save the resulting dataframe back in `df1` to actually change `df1`.

```python
# Your answer here
df1 = df1.dropna(how = 'any', axis = 0)
df1
```

3\. Drop the rows in `df2` in which all values are null. Make sure to save the resulting dataframe back in `df2` to actually change `df2`.

```python
# Your answer here
df2 = df2.dropna(how = 'all', axis = 0)
df2
```

4\. Fill all the missing values in the `sex` column of `df3` with the string *unknown*.

```python
# Your answer here
df3 = df3.fillna(value ={"sex" : "unknown"})
df3
```

5\. Fill all the missing values in the columns `pclass`, `age`, and `fare` of `df4` with the minimum value of their column. Use the <b>min</b> method for each column. 

```python
# Your answer here
df4 = df4.fillna(value = {"pclass": df.pclass.min(), "age": df.age.min(), "fare" : df.fare.min()})
df4
```

6\. Fill all the missing values in `df5` such that the last non-null observations are propagated backward.

```python
# Your answer here
df5 = df5.fillna(method = 'bfill')
df5
```

## Exercises for Aggregation and Grouping

Let's continue to use the entire Titanic datafame `df`.

```python
df = load_dataset("titanic")
df
```

1\. What is the highest fare?

```python
# Your answer here
df.fare.max()
```

2\. What is the lowest fare?

```python
# Your answer here
df.fare.min()
```

3\. What is the mean age of the female passengers? (Select the female passengers first.)

```python
# Your answer here
df[df.sex == 'female'].age.mean()
```

4\. Select the rows, or passengers, who were younger than 10 years and died. (Put two conditions in a maks.)

```python
# Your answer here
df[(df.age < 10) & (df.alive == 'no')]
```

5\. What are the mean ages for those who survived and who died, respectively? Group the dataframe by `survived` and get the mean age of each group. 

```python
# Your answer here
df.groupby("survived").age.mean()
```

6\. Group the dataframe by `survived` and then by `pclass` and get the mean fare of each group. 

```python
# Your answer here
df.groupby(["survived", "pclass"]).fare.mean()
```

7\. Get a copy of `df` in which the entire dataframe is sorted by `age` and then by `pclass`, both in descending order. 

```python
# Your answer here
df2 = df.copy()
df2.sort_values(by = ["age", "pclass"], ascending = [False, False])
```
