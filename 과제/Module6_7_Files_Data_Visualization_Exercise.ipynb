1\. Creat a list `l` of integers from 1 (inclusive) to 100 (inclusive). Write a CSV file named `ex_output.csv` in the `data` directory. The file contains only one column name `id` in the first row. 3\. Create a CSV file named `ex_titanic.csv` in the `data` directory. The file contains three column names `sex`, `pclass` and `embark_town` in the first row and the values from the corresponding columns of `df` in the following rows. The delimiter between columns is a tab. For example, the first 5 lines in the file should look as follows:

```python
# Your answer here
with open("data/ex_titanic.csv", "w") as fw:
    # Write the header row.
    header = "sex\tpclass\tembark_town\n"
    fw.write(header)
    print(header, end="")            # Print the header row just to check the current status.
    
    # Write the value rows.
    for idx, val in df.iterrows():   # Iterates over the rows as (index, series) pairs
        sex = val.sex
        pclass = val.pclass
        embark_town = val.embark_town
        
        row = "{}\t{}\t{}\n".format(sex, pclass, embark_town)
        fw.write(row)
        print(row, end="")           # Print each row to check the current status.
```

1\. Add two new columns `polarity`, which is the polarity score of `text`, and `subjectivity`, which is the subjectivity score of `text`, to `dft` using TextBlob.

```python
# Your answer here
from textblob import TextBlob 

dft["polarity"] = dft.text.apply(lambda x: TextBlob(x).sentiment.polarity)
dft["subjectivity"] = dft.text.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

5\. For the `is_retweet` column  in `dft`, the value 0 means the tweet is a normal tweet, while 1 means the tweet is a retweet. Draw a pie plot that displays the counts of normal tweets and retweets. Set the `title` of the plot to *Normal Tweets vs. Retweets*, the `figsize` to 5 x 5, the `autopct` to '%.1f', and the `fontsize` to 13. Before drawing the pie plot, you may need to calculate the numbers of 0's and 1's in `is_retweet` using the <b>value_counts</b> method.

```python
# Your answer here
count = dft.is_retweet.value_counts(ascending=False)

count.plot(kind="pie", title="Normal Tweets vs. Retweets", figsize=(5,5), autopct='%.1f', fontsize=13)
```

6\. Draw an interactive plot in which a select widget (refer to the documentation below) allows users to choose a tweet type among Normal Tweets, Retweets, and All, and then interactively displays a box plot that displays the distributions of `polarity` and `subjectivity` of the rows with the chosen tweet type. For example, if the user chooses Normal Tweets, it displays a box plot with the distributions of the two columns of the rows with normal tweets. If the user chooses All, it displays a box plot with the distributions of the two columns of all rows. The default value for the widget is All. Set the `figsize` for the box plot to 10 x 5.

Select Widget: https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#Select

```python
from ipywidgets import widgets, interactive, Layout
```

```python
# Your answer here
w_tweet = widgets.Select(
    options=['Normal Tweets', 'Retweets', 'All'],
    value='All',
    description='tweet type',
    disabled=False
)


count = dft.is_retweet.value_counts(ascending=False)
dft["retweet_name"] = ["Normal Tweets" if s ==0 else "Retweets" for s in dft['is_retweet']]


def view(tweet):
    if w_tweet.value == "All":
        df_tmp = dft
    else:
        df_tmp = dft[dft.retweet_name == w_tweet.value]

    title = "Distribution of {}".format(w_tweet.value)
    df_tmp[["polarity", "subjectivity"]].plot(kind="box", title=title, grid=True, figsize=(10,5))

i = interactive(view, tweet=w_tweet)  # The value of the widget is passed as a parameter value.
display(i)
```
